# Standard Deviation and Variance

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
incomes = np.random.normal(100.0, 20.0, 10000)
```

```python
plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std() # where it starts to fall out sharply away from the mean
```

```python
incomes.var()
```
